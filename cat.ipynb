```python
from googletrans import Translator

translator = Translator(service_urls=['translate.google.com'])
```

```python
def translate_column_name(column_name):
  """Translates a column name from French to English."""
  try:
    translation = translator.translate(column_name, src='fr', dest='en')
    return translation.text
  except AttributeError:
    print(f"Translation failed for: {column_name}")
    return column_name

df.columns = [translate_column_name(col) for col in df.columns]
```
